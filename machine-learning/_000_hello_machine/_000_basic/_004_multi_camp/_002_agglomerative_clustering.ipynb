```
import matplotlib.pyplot as plt
from sklearn import cluster
from sklearn import datasets
```

```
# get data
iris = datasets.load_iris()
data = iris['data']
```

```
# create hierarchical type cluster
# 계층형 클러스터 생성

model = cluster.AgglomerativeClustering(n_clusters=3, linkage='ward')
model.fit(data)
```

```
# 예측
labels = model.labels_
```

```
# data 정의
x_index = 2
y_index = 3

data_x=data[:, x_index]
data_y=data[:, y_index]
```

```
x_label = iris['feature_names'][x_index]
y_label = iris['feature_names'][y_index]
```

```
x_label, y_label
```

```
# 산포도
plt.scatter(data_x[labels==0], data_y[labels==0], c='red', alpha=0.15, s=10, marker='o')
plt.scatter(data_x[labels==1], data_y[labels==1], c='blue', alpha=0.15, s=10, marker='o')
plt.scatter(data_x[labels==2], data_y[labels==2], c='yellow', alpha=0.15, s=10, marker='o')

# 축의 라벨, 타이틀 설정
plt.xlabel(x_label, fontsize=7)
plt.ylabel(y_label, fontsize=7)

plt.title('AgglomerativeClustering(ward)', fontsize=10)

plt.show()
```

```
# draw dendrogram
from scipy.cluster.hierarchy import dendrogram, linkage

Z = linkage(data, 'ward')

plt.figure(figsize=(25, 10))
plt.title("hierarchical clustering dendrogram")
plt.xlabel('iris')
plt.ylabel('distance')

dendrogram(Z, leaf_rotation=90., leaf_font_size=8.)
plt.show()
```
