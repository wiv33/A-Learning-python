**matplotlib로 그래프 확장**  

matplotlib는 파이썬에서 시각화에 많이 사용되는 중요한 도구입니다.  
다양한 그림을 그려볼 수 있도록 합시다.

````
# x data의 생성
x = np.linspace(0.2,10,100)

# subplot을 정의합니다.
# matplotlib를 그릴 때 기본적으로 많이 사용되는 구조입니다.
fig, ax = plt.subplots()

# y = 1/x의 그래프
ax.plot(x, 1/x)
# y = log(x)의 그래프
ax.plot(x, np.log(x))
# 그래프의 가로 세로 비율을 1:1이 되도록 합니다.
ax.set_aspect('equal')
# 참고용 격자를 그립니다.
ax.grid(True, which='both')

# 왼쪽 축을 0의 위치에 그립니다.
ax.spines['left'].set_position('zero')

# 오른쪽 축은 그리지 않습니다.
ax.spines['right'].set_color('none')

# y축을 그립니다.
ax.yaxis.tick_left()
ax.spines['bottom'].set_position('zero')
ax.spines['top'].set_color('none')

ax.xaxis.tick_bottom()

plt.show()
````

다시 reshape으로 2차원 데이터를 만드는 구조가 포함되어 있습니다.  

이런식으로 데이터 형식을 변화시키며 연산하는 이유에 대해서 생각해 봅시다.



````
# matplotlib에서 컬러맵(color map) 임포트
from matplotlib import cm

w = np.array([1.2,1.0])
b = 0
x1 = np.linspace(-5, 5, num=50)
x2 = np.linspace(-5, 5, num=50)

# 각각의 함수 사용후에 변수의 shape과 data를 print하면서 데이터 구조를 파악해 봅시다.
x1, x2 = np.meshgrid(x1, x2)

x = np.vstack((x1.flatten(), x2.flatten())).T

# 함수가 조금 복잡합니다. 다른 간단한 함수로 바꾸어서도 그려봅시다.
y = ((x1-3)**2)*((x1+3)**2) -20*x1+ ((x2-3)**2)*(((x2+3))**2)

# y를 다시 2차원 구조로 변경합니다.
y = y.reshape(len(x1), len(x2))

fig = plt.figure()
ax = fig.gca(projection='3d')

# surface 그래프 그리기
surf = ax.plot_surface(x1, x2, y, cmap=cm.coolwarm,
                       linewidth=0, antialiased=False)
plt.show()
````
